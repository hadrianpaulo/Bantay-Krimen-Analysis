```python
%pylab inline
import pandas as pd
from sklearn.cluster import DBSCAN, MeanShift
import folium
```

```python
df = pd.read_csv('../data/processed/crime_incidences.csv')
df_ncr = df[df.region == 'NCR']
```

```python
dbscan_model = DBSCAN(eps=0.001, n_jobs=-1).fit(df_ncr[['lat', 'lng']])
meanshift_model = MeanShift(n_jobs=-1).fit(df_ncr[['lat', 'lng']])
```

```python
dbscan_crime = {}
meanshift_crime = {}
for crime_type in df_ncr.crime.unique():
    dbscan_crime[crime_type] = DBSCAN(eps=0.001, n_jobs=-1).fit(
        df_ncr[df_ncr.crime == crime_type][['lat', 'lng']])
    meanshift_crime[crime_type] = MeanShift(n_jobs=-1).fit(
        df_ncr[df_ncr.crime == crime_type][['lat', 'lng']])
```

```python
len(dbscan_model.core_sample_indices_)
```

```python
%store meanshift_model
```

```python
%store meanshift_crime
```

```python
m = folium.Map(location=[13.5, 122.4], zoom_start=6)
for location in meanshift_model.cluster_centers_:
    folium.CircleMarker(location).add_to(m)
```
